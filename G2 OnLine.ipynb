```python
def debar(elm):
    return d(elm) - de(elm)
```

```python
def baro(elem):
    if elem in CC:
        return elem.conjugate()
    return sum([elem.interior_product(c).constant_coefficient().conjugate()*c for c in E.basis()])

def hodge_star(elem):
    return baro(elem.hodge_dual())
```

```python
def laplac_debar(elem):
    return debar(hodge_star(debar(hodge_star(elem)))) + hodge_star(debar(hodge_star(debar(elem))))
```

```python
def laplac_A(elem):
    return de(hodge_star(de(hodge_star(elem)))) + debar(hodge_star(debar(hodge_star(elem)))) +\
hodge_star(de(debar(hodge_star(de(debar(elem)))))) + de(debar(hodge_star(de(debar(hodge_star(elem)))))) +\
hodge_star(debar(hodge_star(de(hodge_star(hodge_star(debar(hodge_star(de(hodge_star(elem)))))))))) + \
debar(hodge_star(de(hodge_star(hodge_star(debar(hodge_star(de(hodge_star(hodge_star(elem))))))))))
```

```python
def laplac_BC(elem):
    return hodge_star(de(debar(hodge_star(de(debar(elem)))))) + de(debar(hodge_star(de(debar(hodge_star(elem)))))) +\
        hodge_star(debar(hodge_star(de(hodge_star(hodge_star(debar(hodge_star(de(hodge_star(elem))))))))))+\
        hodge_star(hodge_star(debar(hodge_star(de(hodge_star(hodge_star(debar(hodge_star(de(elem))))))))))+\
        hodge_star(debar(hodge_star(debar(elem)))) + hodge_star(de(hodge_star(de(elem))))
```
